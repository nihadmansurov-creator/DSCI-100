## Description of the dataset



The players.csv dataset has 196 rows and 7 columns, describing player characteristics like name, gender ,and age. Some players are more advanced than the others ,so they are placed into different experience categories. Some are pro, meaning they are quite advanced in playing the game, while others are still beginners.The played_hours column describes how many hours each player has spent playing the game.The dataset shows that some individuals have spent significantly more time playing the game than others.Additionally,the subscribe column shows whether the player is subscribed to the newsletter.Lastly, the hashed and protected versions of players' email adresses are stored in the hashedEmail column.

### collection of data:

Some of the data for player.csv dataset was most probably collected during the registration, when players had to create game accounts using their names , ages, genders.However, experience was likely determined from the way individuals played the game.The information for played_hours column was probably collected by recording how long each player was in the game.

## issues :


- Some experience categories in players.csv dataset may have far more players than the others, which can lead to bias results when doing calculations.Specifically, it can lead to less reliable comparisons between the experience categories.
- The size of the sample is small. We only have 196 rows ,meaning only 196 players. If the sample size would be greater , there would be more accurate results.

My choice the  question 2 : "We would like to know which kinds of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts".

My specific question is "Do players with higher experience level spend more time on the game than the players with lower experience levels.So i will use the experience and played_hours from my players.csv to compare how much time individuals in different experience categories spend playing the game. I will keep the required columns and non-zero rows.I will convert experience column into categorical factor, so later i can plot linear regression to check whether the experience level predicts played_hours.

### mean for each variable:


```r
mean_table <- players |>
summarise(mean_played_hours = round(mean(played_hours, na.rm = TRUE), 2),
         mean_age = round(mean(Age, na.rm = TRUE), 2))
mean_table
```

## tidying data:



```r
players_tidy <- players |>
                    select(experience, played_hours) 
                    
players_tidy
```

## mean for each category:

```r
players_mean_cat <- players_tidy |>
                    group_by(experience) |>
                    summarise(players_mean_cat = round(mean(played_hours, na.rm = TRUE), 2))
players_mean_cat




                    
```

```r
## mean played_hours for each category 

experience_new <- players |>
                    mutate(experience = as.factor(experience)) 
mean_players_plot <- players_mean_cat |>
                    ggplot(aes(x = experience, y = players_mean_cat, fill = experience)) +
                    geom_col() +
                    labs(title =  "Mean played hours for each experience level" ,
                     x = "Experience level",
                     y = "Played hours (hours)") +
                     theme_minimal() 



mean_players_plot
```

```r
## total played hours per each category 

total_played_hours <- players |>
                    group_by(experience) |>
                    summarise(total_played_hours = sum(played_hours, na.rm = TRUE))
total_played_hours
players_plot <-total_played_hours |>
                    ggplot(aes(x = experience, y = total_played_hours, fill = experience)) +
                    geom_col() +
                    labs(title =  "Total played hours for each experience level" ,
                     x = "Experience level",
                     y = "Played hours (hours)") +
                     theme_minimal() 
players_plot
```

From the first visualization we directly see if the experience level corresponds to a longer gameplay time.The second plot shows how much overall gameplay each experience group contributes, which shows the most active groups.
